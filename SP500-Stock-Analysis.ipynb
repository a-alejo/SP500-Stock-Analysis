```python
# Create a copy of the 'stocks' DataFrame
stocks_flatten = stocks.copy()

# Convert multi-index into tuple incase needed
stocks_flatten.columns = stocks_flatten.columns.to_flat_index()

# preview changes
stocks_flatten.head()
```

```python
close = stocks["Close"]
close
```

```python
# Load visualization libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Set the width and height of the figure
plt.figure(figsize=(15, 5))

# Create a line plot
sns.lineplot(data=close, dashes=False)
```

```python
# Looking at the stocks from 2015

close_reset = close.copy()

close_reset = close_reset.reset_index()
close_reset = close_reset[close_reset["Date"] > "2015"]
close_reset.set_index("Date", inplace=True)


# Create a line plot
sns.lineplot(data=close_reset, dashes=False)
plt.title("Prices after 2015")
plt.ylabel("Stock Price")
```

```python
# Normalize the data by multiplying close price by 100
close_normalized = close_reset.copy()
close_normalized = close_normalized.div(close_normalized.iloc[0]).mul(100)

# Set the width and height of the figure
plt.figure(figsize=(15, 5))

# Create a line plot
sns.lineplot(data=close_normalized, dashes=False)
plt.title("Normalized Stock Price")
plt.ylabel("Stock Price")
```

```python
# Create a new dataframe containing Apple's close price
apple_close = apple["Close"].copy().to_frame()

# Get price difference between a day
apple_close["price_diff"] = apple_close["Close"].diff()

# Calculate the percentage difference between a day
apple_close["percent_diff"] = apple_close["Close"].pct_change(periods=1).mul(100)

# Preview table
apple_close
```

```python
# Get the last business day of each month
apple_close.resample("bm").last()
```

```python
# Plot the trading returns range
sns.histplot(data=apple_close["percent_diff"].resample("bm").last())
plt.title("Returns Percentage Range")

# Zoom in plot to hide outliers
plt.xlim(-10, 10)
```

```python
close.describe().T  # [['mean','std']]
```

```python
# Separate the mean and std
closed = close.describe().T[["mean", "std"]]

for index, row in closed.iterrows():
    x = row["std"]
    y = row["mean"]
    plt.scatter(x, y)
    plt.annotate(index, (x, y))


plt.xlabel("Annual Risk (STD)")
plt.ylabel("Annual Return (Mean)")
plt.title("Risk vs Reward")
```

As we see above on the scatter plot - Of the three stocks Apple is the "safer" option with relative low risk to return, Amazon has a much higher risk but a low return rate and on the far right side Microsoft offers a higher return to risk when comparing the three
